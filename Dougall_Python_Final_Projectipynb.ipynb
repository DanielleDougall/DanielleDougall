The full script I used to analyze all the YouTube comments for the GOP‚Äôs ‚ÄúBeat Biden‚Äù ad can be found at the end of this page. I used ChatGPT to help me generate the script to import the YouTube comments and for problem-solving with code errors.



```python
import googleapiclient.discovery
import pandas as pd

API_KEY = 'AIzaSyApgu78mpZPbFmcWTkbKnnmTRe7HO-ruig'
VIDEO_ID = 'kLMMxgtxQ1Y'

def extract_all_comments(video_id, api_key, max_results=1274):
    youtube = googleapiclient.discovery.build('youtube', 'v3', developerKey=api_key)

    comments = []

    # Fetch top-level comments
    nextPageToken = None
    while len(comments) < max_results:
        response = youtube.commentThreads().list(
            part="snippet",
            videoId=video_id,
            maxResults=100,
            textFormat="plainText",
            pageToken=nextPageToken
        ).execute()

        for item in response["items"]:
            comment_text = item["snippet"]["topLevelComment"]["snippet"]["textDisplay"]
            if isinstance(comment_text, str):
                comments.append(comment_text)

            # Fetch replies for the current top-level comment
            if item["snippet"]["totalReplyCount"] > 0:
                reply_nextPageToken = None
                while reply_nextPageToken or not reply_nextPageToken:
                    reply_response = youtube.comments().list(
                        part="snippet",
                        parentId=item["id"],
                        maxResults=100,
                        textFormat="plainText",
                        pageToken=reply_nextPageToken
                    ).execute()
                    for reply in reply_response["items"]:
                        reply_text = reply["snippet"]["textDisplay"]
                        if isinstance(reply_text, str):
                            comments.append(reply_text)
                    reply_nextPageToken = reply_response.get('nextPageToken')
                    if not reply_nextPageToken:
                        break

        nextPageToken = response.get('nextPageToken')
        if not nextPageToken:
            break

    return comments[:max_results]

comments = extract_all_comments(VIDEO_ID, API_KEY, max_results=1274)

# Create the DataFrame
comments_df = pd.DataFrame({
    'comment_number': range(1, len(comments) + 1),
    'comment': comments
})

# Display the DataFrame to ensure it's formatted correctly
print(comments_df)
```

### **Removing Stop Words**

This defines a function to remove common stopwords from the comments. Stop words are filtered out to focus on the more meaningful words in sentiment analysis.

```python
# Import necessary libraries
import pandas as pd
import nltk
from nltk.corpus import stopwords

# Download stopwords if not already downloaded
nltk.download('stopwords')

# Define a function to clean a comment by removing stop words
stop_words = set(stopwords.words('english'))
def clean_comment(comment):
    words = comment.split()
    clean_words = [word for word in words if word.lower() not in stop_words]
    return ' '.join(clean_words)

# Create a DataFrame
comments_df = pd.DataFrame({'comment': ["This is a sample comment with stop words.",
                                        "Another comment to clean."]})

# Convert all comments to lowercase before cleaning
comments_df['clean_comment'] = comments_df['comment'].str.lower()

# Remove stop words from the 'clean_comment' column
comments_df['clean_comment'] = comments_df['clean_comment'].apply(clean_comment)

# Display the result
print(comments_df)

```
